# 3. Cleaning
## 3.1 Label encoding

```python
labelencoder = LabelEncoder()
data["species"] = labelencoder.fit_transform(data["species"])
```

```python
data.head()
```

## 3.2 Renommer "species" en "target"

```python
data = data.rename(columns={"species": "target"})
data.head()
```

# 4. Export csv

```python
path = "data/cleaned/"
filename = "data_cleaned.csv"
```

```python
data.to_csv(path+filename, index=False)
```
